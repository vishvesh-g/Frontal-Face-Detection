```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = plt.imread("Group Photo1.jpeg")
```

```python
img.shape
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt (1).xml")
```

```python
faces = detector.detectMultiScale(img)
```

```python
##all Front facing faces identified
faces
```

```python
## this give idea of what 'faces' stores that is starting x, y coordinates and width, height respectively
## x,y,w,h = faces[0]
```

```python
## this then modifies the image bu adding rectange starting form (x,y) to (x+w,y+h) of desired colour, here it is (255,255,255) i.e. white
## modified_img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255), 2)
```

```python
## plt.imshow(modified_img)
```

```python
## this is for marking all frontal faces white
for face in faces:
    x, y, w, h = face
    img = cv2.rectangle(img, (x,y), (x+w, y+h), (255,255,255), 3)
    
```

```python
plt.imshow(img)
```
